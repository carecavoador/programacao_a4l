```python
import re

from pypdf import PdfReader, PageObject

from item import Item


r = PdfReader("MPOP800.PDF")
paginas = r.pages
```

```python
def coleta_itens(pagina: PageObject) -> list:
    itens = []
    for l in pagina:
        # localizar a coluna OP's
        inicio_item = re.findall(r'\d{3}\.\d{3}', l)
        if inicio_item:
            # Começa a buscar os dados de cada item
            ops = inicio_item[0]
            idx = pg_1.index(ops)
            descricao_de_produto = pg_1[idx + 1]
            apr_rel_prod = pg_1[idx + 2]
            qtd_prev_m = pg_1[idx + 3]
            formato = pg_1[idx + 5]
            faca = pg_1[idx + 7]
            cilin = pg_1[idx + 8]
            sit = pg_1[idx + 9]
            
            # Verifica se tem Hot-S na próxima coluna. Caso tenha, incrementa o Index
            tem_hot_s = pg_1[idx + 10].startswith('Nº')
            hot_s = None
            if tem_hot_s:
                hot_s = pg_1[idx + 10]
                idx += 1

            qtf_prev_mil = pg_1[idx + 10].replace(',', '.')
            sai = pg_1[idx + 11]
            alm = pg_1[idx + 12]
            pre = pg_1[idx + 13]
            cli = pg_1[idx + 14]
            col = pg_1[idx + 16]
            des = pg_1[idx + 17]
            fornec = pg_1[idx + 18]
            cart = pg_1[idx + 19]
            entrega = pg_1[idx + 21]
            cod_sap = pg_1[idx + 22]

            # Verifica se tem Roto na próxima coluna. Caso tenha, incrementa o Index
            tem_roto = pg_1[idx + 23].startswith('n°')
            roto = None
            if tem_roto:
                roto = pg_1[idx + 23]
                idx += 1

            prv_ini = pg_1[idx + 23]
            setup = pg_1[idx + 24]
            acerto = pg_1[idx + 25]
            oper = pg_1[idx + 26]
            bula = pg_1[idx + 27]
            mont = pg_1[idx + 29]


            itens.append(
                Item(
                    ops=ops,
                    descricao_de_produto=descricao_de_produto,
                    apr_rel_prod=apr_rel_prod,
                    qtd_prev_m=qtd_prev_m,
                    formato=formato,
                    faca=faca,
                    cilin=cilin,
                    sit=sit,
                    qtf_prev_mil=qtf_prev_mil,
                    sai=sai,
                    alm=alm,
                    pre=pre,
                    cli=cli,
                    col=col,
                    des=des,
                    fornec=fornec,
                    cart=cart,
                    entrega=entrega,
                    cod_sap=cod_sap,
                    prv_ini=prv_ini,
                    setup=setup,
                    acerto=acerto,
                    oper=oper,
                    bula=bula,
                    mont=mont,
                    hot_s=hot_s,
                    roto=roto
                )
            )
    return itens
```

```python
texto_paginas = [pg.extract_text() for pg in paginas]
pg_1 = texto_paginas[0].splitlines()
pg_1 = [l.strip() for l in pg_1]

for item in coleta_itens(pg_1):
    print(str(item))
```
