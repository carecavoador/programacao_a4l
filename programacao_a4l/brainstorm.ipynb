```python
import re

from pypdf import PdfReader, PageObject

from item import Item


r = PdfReader("MPOP800.PDF")
paginas = r.pages
```

```python
def coleta_itens(pagina: list[str]) -> list:
    itens = []
    for l in pagina:
        # localizar a coluna OP's
        inicio_item = re.findall(r'\d{3}\.\d{3}', l)
        if inicio_item:
            # Começa a buscar os dados de cada item
            ops = inicio_item[0]
            try:
                idx = pagina.index(ops)
            except ValueError:
                print('Item não encontrado:', ops)
            descricao_de_produto = pagina[idx + 1]
            apr_rel_prod = pagina[idx + 2]
            qtd_prev_m = pagina[idx + 3]
            formato = pagina[idx + 5]
            faca = pagina[idx + 7]
            cilin = pagina[idx + 8]
            sit = pagina[idx + 9]
            
            # Verifica se tem Hot-S na próxima coluna. Caso tenha, incrementa o Index
            tem_hot_s = pagina[idx + 10].startswith('Nº')
            hot_s = None
            if tem_hot_s:
                hot_s = pagina[idx + 10]
                idx += 1

            qtf_prev_mil = pagina[idx + 10].replace(',', '.')
            sai = pagina[idx + 11]
            alm = pagina[idx + 12]
            pre = pagina[idx + 13]
            cli = pagina[idx + 14]
            col = pagina[idx + 16]
            des = pagina[idx + 17]
            fornec = pagina[idx + 18]
            cart = pagina[idx + 19]
            entrega = pagina[idx + 21]
            cod_sap = pagina[idx + 22]

            # Verifica se tem Roto na próxima coluna. Caso tenha, incrementa o Index
            tem_roto = pagina[idx + 23].startswith('n°')
            roto = None
            if tem_roto:
                roto = pagina[idx + 23]
                idx += 1

            prv_ini = pagina[idx + 23]
            setup = pagina[idx + 24]
            acerto = pagina[idx + 25]
            oper = pagina[idx + 26]
            bula = pagina[idx + 27]
            mont = pagina[idx + 29]

            itens.append(
                Item(
                    ops=ops,
                    descricao_de_produto=descricao_de_produto,
                    apr_rel_prod=apr_rel_prod,
                    qtd_prev_m=qtd_prev_m,
                    formato=formato,
                    faca=faca,
                    cilin=cilin,
                    sit=sit,
                    qtf_prev_mil=qtf_prev_mil,
                    sai=sai,
                    alm=alm,
                    pre=pre,
                    cli=cli,
                    col=col,
                    des=des,
                    fornec=fornec,
                    cart=cart,
                    entrega=entrega,
                    cod_sap=cod_sap,
                    prv_ini=prv_ini,
                    setup=setup,
                    acerto=acerto,
                    oper=oper,
                    bula=bula,
                    mont=mont,
                    hot_s=hot_s,
                    roto=roto
                )
            )
    return itens
```

```python
todos_itens = []

for pagina in paginas:
    texto_pagina = pagina.extract_text().splitlines()
    texto_pagina = [linha.strip() for linha in texto_pagina]
    # print(texto_pagina)
    itens_pagina = coleta_itens(texto_pagina)
    todos_itens.extend(itens_pagina)

for item in todos_itens:
    print(str(item))

# texto_paginas = [pg.extract_text() for pg in paginas]
# pg_1 = texto_paginas[0].splitlines()
# pg_1 = [l.strip() for l in pg_1]

# for item in coleta_itens(pg_1):
#     print(str(item))
```
